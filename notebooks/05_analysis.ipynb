# 05 — Analysis & interpretability (Colab driver)

Post-hoc, on a trained checkpoint. Thin driver over `nmt.analysis`.

**Steps:**
1. `alignments` — learned cross-attention vs `fast_align`/`eflomal` oracle; AER; best-aligned head.
   (Install the C++ aligner *here*, not in pyproject; degrade gracefully if absent.)
2. `attention_viz` — per-head heatmaps (encoder self, decoder self, cross).
3. `head_importance` — head ablation/importance (Michel et al.) + prune-and-eval.
4. `embeddings` — t-SNE / UMAP of the tied embeddings; clustering sanity.
5. `probing` — POS/length/NER classifiers on frozen (detached) encoder states.
6. `loss_breakdown` — per-token loss histogram; loss vs source length / complexity buckets.

Needs the model's `return_attn=True` path.
