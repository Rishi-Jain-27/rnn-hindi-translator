# 03 — Back-translation augmentation (Colab driver)

Runs **after** a working forward model. Thin driver over `nmt.data.backtranslation` + `nmt.data.monolingual`.

**Steps:**
1. Train a reverse **en→hi** model — run **`02b_train_reverse`** (`02_train` with source/target swapped; checkpoints land in `checkpoints/backtranslate`).
2. `monolingual` — stage + dedupe + dev/test-leak-filter monolingual English (on Drive).
3. `backtranslation` — sampled decode of mono-EN → synthetic Hindi; prefix `<bt>` tag (tagged BT).
4. Merge synthetic pairs into the forward training set; re-run `02_train`.

Note: Steps 1 and 4 are each a full training run — back-translation is the biggest BLEU lever after data scale, but Colab-credit heavy.
