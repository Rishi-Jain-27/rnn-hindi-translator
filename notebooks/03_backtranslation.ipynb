# 03 ‚Äî Back-translation augmentation (Colab driver)

Runs **after** a working forward model. Thin driver over `nmt.data.backtranslation` + `nmt.data.monolingual`.

**Steps:**
1. Train a reverse **en‚Üíhi** model ‚Äî reuse the whole stack with src/tgt swapped (`02_train` config, flipped).
2. `monolingual` ‚Äî stage + dedupe + dev/test-leak-filter monolingual English (on Drive).
3. `backtranslation` ‚Äî sampled decode of mono-EN ‚Üí synthetic Hindi; prefix `<bt>` tag (tagged BT).
4. Merge synthetic pairs into the forward training set; re-run `02_train`.

Note: this is a full second training run ‚Äî biggest BLEU lever after data scale, but Colab-credit heavy.
